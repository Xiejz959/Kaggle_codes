```python
valid_start = pd.Timestamp("2017-07-31")
valid_end = pd.Timestamp("2017-08-15")

train_part = train[train["date"] < valid_start].copy()
valid_part = train[(train["date"] >= valid_start) & (train["date"] <= valid_end)].copy()

print("train_part:", train_part["date"].min(), "to", train_part["date"].max(), train_part.shape)
print("valid_part:", valid_part["date"].min(), "to", valid_part["date"].max(), valid_part.shape)
print("test      :", test["date"].min(), "to", test["date"].max(), test.shape)
```

```python
def rmsle(y_true, y_pred):
    y_pred = np.clip(y_pred, 0, None)
    return np.sqrt(np.mean((np.log1p(y_pred) - np.log1p(y_true)) ** 2))
```

```python
baseline_mean = (
    train_part
    .groupby(["store_nbr", "family"])["sales"]
    .mean()
    .reset_index()
    .rename(columns={"sales": "baseline_pred"})
)

valid_baseline = valid_part.merge(
    baseline_mean,
    on=["store_nbr", "family"],
    how="left"
)

global_mean = train_part["sales"].mean()
valid_baseline["baseline_pred"] = valid_baseline["baseline_pred"].fillna(global_mean)

baseline_score = rmsle(valid_baseline["sales"], valid_baseline["baseline_pred"])
print("Baseline RMSLE:", baseline_score)
```

```python
recent_start = valid_start - pd.Timedelta(days=28)

recent_mean = (
    train_part[train_part["date"] >= recent_start]
    .groupby(["store_nbr", "family"])["sales"]
    .mean()
    .reset_index()
    .rename(columns={"sales": "recent_28_mean"})
)

valid_recent = valid_part.merge(
    recent_mean,
    on=["store_nbr", "family"],
    how="left"
)

valid_recent["recent_28_mean"] = valid_recent["recent_28_mean"].fillna(global_mean)

recent_score = rmsle(valid_recent["sales"], valid_recent["recent_28_mean"])
print("Recent 28-day mean RMSLE:", recent_score)
```

```python
train_part = train_part.copy()
valid_part = valid_part.copy()

train_part["dayofweek"] = train_part["date"].dt.dayofweek
valid_part["dayofweek"] = valid_part["date"].dt.dayofweek

dow_mean = (
    train_part[train_part["date"] >= recent_start]
    .groupby(["store_nbr", "family", "dayofweek"])["sales"]
    .mean()
    .reset_index()
    .rename(columns={"sales": "recent_28_dow_mean"})
)

valid_dow = valid_part.merge(
    dow_mean,
    on=["store_nbr", "family", "dayofweek"],
    how="left"
)

valid_dow["recent_28_dow_mean"] = valid_dow["recent_28_dow_mean"].fillna(global_mean)

dow_score = rmsle(valid_dow["sales"], valid_dow["recent_28_dow_mean"])
print("Recent 28-day + dayofweek RMSLE:", dow_score)
```

```python
def same_month_day_baseline(valid_df, history_df):
    history_same_day = history_df.copy()
    valid_same_day = valid_df.copy()

    history_same_day["month"] = history_same_day["date"].dt.month
    history_same_day["day"] = history_same_day["date"].dt.day

    valid_same_day["month"] = valid_same_day["date"].dt.month
    valid_same_day["day"] = valid_same_day["date"].dt.day

    same_day_mean = (
        history_same_day
        .groupby(["store_nbr", "family", "month", "day"])["sales"]
        .mean()
        .reset_index()
        .rename(columns={"sales": "same_month_day_mean"})
    )

    valid_same_day = valid_same_day.merge(
        same_day_mean,
        on=["store_nbr", "family", "month", "day"],
        how="left",
    )

    valid_same_day["same_month_day_mean"] = valid_same_day["same_month_day_mean"].fillna(global_mean)

    score = rmsle(
        valid_same_day["sales"],
        valid_same_day["same_month_day_mean"]
    )

    return score, valid_same_day["same_month_day_mean"]

same_day_score, same_day_pred = same_month_day_baseline(valid_part, train_part)

print("Same month-day historical mean RMSLE:", same_day_score)
```

```python
def lag_baseline(valid_df, history_df, lag_days):
    valid_lag = valid_df.copy()
    valid_lag["lookup_date"] = valid_lag["date"] - pd.Timedelta(days=lag_days)

    lag_ref = (
        history_df[["date", "store_nbr", "family", "sales"]]
        .rename(columns={
            "date": "lookup_date",
            "sales": f"lag_{lag_days}",
        })
    )

    valid_lag = valid_lag.merge(
        lag_ref,
        on=["lookup_date", "store_nbr", "family"],
        how="left",
    )

    pred_col = f"lag_{lag_days}"
    valid_lag[pred_col] = valid_lag[pred_col].fillna(global_mean)

    score = rmsle(valid_lag["sales"], valid_lag[pred_col])
    return score, valid_lag[pred_col]
```

```python
lag_scores = {}
lag_predictions = {}

for lag in [7, 14, 28]:
    score, pred = lag_baseline(valid_part, train_part, lag)
    lag_scores[f"lag_{lag}"] = score
    lag_predictions[f"lag_{lag}"] = pred
    print(f"lag_{lag} RMSLE:", score)
```

```python
valid_lag_avg = valid_part.copy()
valid_lag_avg["pred"] = np.mean(
    [lag_predictions["lag_7"], lag_predictions["lag_14"], lag_predictions["lag_28"]],
    axis=0,
)

lag_avg_score = rmsle(valid_lag_avg["sales"], valid_lag_avg["pred"])

print("Average lag 7/14/28 RMSLE:", lag_avg_score)
```

```python
train_fe = train.copy()
test_fe = test.copy()

train_fe["is_train"] = 1
test_fe["is_train"] = 0
test_fe["sales"] = np.nan

all_data = pd.concat([train_fe, test_fe], axis=0, ignore_index=True)
all_data = all_data.sort_values(["store_nbr", "family", "date"]).reset_index(drop=True)
```

```python
def add_date_features(df):
    df = df.copy()
    df["year"] = df["date"].dt.year
    df["month"] = df["date"].dt.month
    df["day"] = df["date"].dt.day
    df["dayofweek"] = df["date"].dt.dayofweek
    df["weekofyear"] = df["date"].dt.isocalendar().week.astype(int)
    df["is_weekend"] = df["dayofweek"].isin([5, 6]).astype(int)
    df["is_month_start"] = df["date"].dt.is_month_start.astype(int)
    df["is_month_end"] = df["date"].dt.is_month_end.astype(int)
    return df

all_data = add_date_features(all_data)
```

```python
oil_fe = oil.sort_values("date").copy()
oil_fe["dcoilwtico"] = oil_fe["dcoilwtico"].ffill().bfill()

all_data = all_data.merge(stores, on="store_nbr", how="left")
all_data = all_data.merge(oil_fe, on="date", how="left")
```

```python
holidays_fe = holidays[holidays["transferred"] == False].copy()

holiday_simple = (
    holidays_fe
    .groupby("date")
    .agg(
        is_holiday=("date", "size"),
        holiday_type=("type", "first"),
        holiday_locale=("locale", "first"),
    )
    .reset_index()
)

holiday_simple["is_holiday"] = 1

all_data = all_data.merge(holiday_simple, on="date", how="left")

all_data["is_holiday"] = all_data["is_holiday"].fillna(0).astype(int)
all_data["holiday_type"] = all_data["holiday_type"].fillna("None")
all_data["holiday_locale"] = all_data["holiday_locale"].fillna("None")
```

```python
group_cols = ["store_nbr", "family"]

for lag in [1, 7, 14, 28]:
    all_data[f"lag_{lag}"] = (
        all_data
        .groupby(group_cols)["sales"]
        .shift(lag)
    )

for window in [7, 14, 28]:
    all_data[f"rolling_mean_{window}"] = (
        all_data
        .groupby(group_cols)["sales"]
        .transform(lambda s: s.shift(1).rolling(window).mean())
    )

    all_data[f"rolling_std_{window}"] = (
        all_data
        .groupby(group_cols)["sales"]
        .transform(lambda s: s.shift(1).rolling(window).std())
    )
```

```python
eps = 1e-6

all_data["trend_7_28"] = all_data["rolling_mean_7"] / (all_data["rolling_mean_28"] + eps)
all_data["trend_14_28"] = all_data["rolling_mean_14"] / (all_data["rolling_mean_28"] + eps)

valid_start = pd.Timestamp("2017-07-31")

feature_data = all_data.copy()

train_model = feature_data[
    (feature_data["is_train"] == 1) &
    (feature_data["date"] < valid_start)
].copy()

valid_model = feature_data[
    (feature_data["is_train"] == 1) &
    (feature_data["date"] >= valid_start)
].copy()

test_model = feature_data[
    feature_data["is_train"] == 0
].copy()

print(train_model.shape, valid_model.shape, test_model.shape)

target_col = "sales"

drop_cols = [
    "id",
    "date",
    "sales",
    "is_train",
]

feature_cols = [col for col in train_model.columns if col not in drop_cols]

cat_cols = train_model[feature_cols].select_dtypes(include=["object", "string"]).columns.tolist()
num_cols = [col for col in feature_cols if col not in cat_cols]


print("categorical:", cat_cols)
print("numeric:", num_cols)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
```

```python
cat_cols = train_model[feature_cols].select_dtypes(include=["object", "string"]).columns.tolist()
num_cols = [col for col in feature_cols if col not in cat_cols]

X_train = train_model[feature_cols]
y_train = np.log1p(train_model["sales"])

X_valid = valid_model[feature_cols]
y_valid = valid_model["sales"]

print("X_train:", X_train.shape)
print("X_valid:", X_valid.shape)
print("categorical:", cat_cols)
print("numeric:", len(num_cols))
```

```python
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        (
            "encoder",
            OrdinalEncoder(
                handle_unknown="use_encoded_value",
                unknown_value=-1,
            ),
        ),
    ]
)

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ]
)

X_train_processed = preprocess.fit_transform(X_train)
X_valid_processed = preprocess.transform(X_valid)

print("processed train:", X_train_processed.shape)
print("processed valid:", X_valid_processed.shape)
```

```python
model = HistGradientBoostingRegressor(
    max_iter=1,
    learning_rate=0.06,
    max_leaf_nodes=31,
    l2_regularization=0.1,
    random_state=42,
    warm_start=True,
)

total_iter = 300
step = 10

for n_iter in tqdm(range(step, total_iter + step, step)):
    model.set_params(max_iter=n_iter)
    model.fit(X_train_processed, y_train)

valid_pred_log = model.predict(X_valid_processed)
valid_pred = np.expm1(valid_pred_log)
valid_pred = np.clip(valid_pred, 0, None)

score = rmsle(y_valid, valid_pred)
mae = mean_absolute_error(y_valid, valid_pred)

print("Model RMSLE:", score)
print("Model MAE:", mae)
```

```python
valid_result = valid_model[["id", "date", "store_nbr", "family", "sales"]].copy()
valid_result["pred"] = valid_pred
valid_result["abs_error"] = np.abs(valid_result["sales"] - valid_result["pred"])
valid_result["log_error"] = np.abs(np.log1p(valid_result["sales"]) - np.log1p(valid_result["pred"]))

display(valid_result.head())
```

```python
family_error = (
    valid_result
    .groupby("family")
    .agg(
        rmsle=("log_error", lambda x: np.sqrt(np.mean(x ** 2))),
        mae=("abs_error", "mean"),
        actual_mean=("sales", "mean"),
        pred_mean=("pred", "mean"),
        count=("sales", "size"),
    )
    .sort_values("rmsle", ascending=False)
)

display(family_error)
```

```python
store_error = (
    valid_result
    .groupby("store_nbr")
    .agg(
        rmsle=("log_error", lambda x: np.sqrt(np.mean(x ** 2))),
        mae=("abs_error", "mean"),
        actual_mean=("sales", "mean"),
        pred_mean=("pred", "mean"),
        count=("sales", "size"),
    )
    .sort_values("rmsle", ascending=False)
)

display(store_error.head(20))
```

```python
date_error = (
    valid_result
    .groupby("date")
    .agg(
        rmsle=("log_error", lambda x: np.sqrt(np.mean(x ** 2))),
        mae=("abs_error", "mean"),
        actual_total=("sales", "sum"),
        pred_total=("pred", "sum"),
    )
    .reset_index()
)

display(date_error)

plt.figure(figsize=(14, 4))
plt.plot(date_error["date"], date_error["actual_total"], label="Actual")
plt.plot(date_error["date"], date_error["pred_total"], label="Pred")
plt.title("Validation Actual vs Predicted Total Sales")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=valid_result.sample(min(20000, len(valid_result)), random_state=42),
    x="sales",
    y="pred",
    alpha=0.3,
)
plt.plot([0, valid_result["sales"].max()], [0, valid_result["sales"].max()], color="red")
plt.title("Actual vs Predicted Sales")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
total_actual = valid_result["sales"].sum()
total_pred = valid_result["pred"].sum()

print("actual total:", total_actual)
print("pred total  :", total_pred)
print("pred / actual:", total_pred / total_actual)
print("actual / pred:", total_actual / total_pred)
```

```python
date_bias = (
    valid_result
    .groupby("date")
    .agg(
        actual_total=("sales", "sum"),
        pred_total=("pred", "sum"),
    )
    .reset_index()
)

date_bias["pred_actual_ratio"] = date_bias["pred_total"] / date_bias["actual_total"]

display(date_bias)

plt.figure(figsize=(14, 4))
plt.plot(date_bias["date"], date_bias["pred_actual_ratio"], marker="o")
plt.axhline(1.0, color="red", linestyle="--")
plt.title("Prediction / Actual Ratio by Date")
plt.xlabel("Date")
plt.ylabel("Pred / Actual")
plt.show()
```

```python
family_bias = (
    valid_result
    .groupby("family")
    .agg(
        actual_mean=("sales", "mean"),
        pred_mean=("pred", "mean"),
        actual_total=("sales", "sum"),
        pred_total=("pred", "sum"),
    )
    .reset_index()
)

family_bias["pred_actual_ratio"] = family_bias["pred_total"] / family_bias["actual_total"]
family_bias = family_bias.sort_values("pred_actual_ratio")

display(family_bias)
```

```python
for start_date in ["2015-08-01", "2016-01-01", "2016-08-01"]:
    train_tmp = train_model[train_model["date"] >= pd.Timestamp(start_date)].copy()

    X_train_tmp = train_tmp[feature_cols]
    y_train_tmp = np.log1p(train_tmp["sales"])

    preprocess_tmp = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, num_cols),
            ("cat", categorical_transformer, cat_cols),
        ]
    )

    X_train_tmp_processed = preprocess_tmp.fit_transform(X_train_tmp)
    X_valid_tmp_processed = preprocess_tmp.transform(X_valid)

    model_tmp = HistGradientBoostingRegressor(
        max_iter=200,
        learning_rate=0.06,
        max_leaf_nodes=31,
        l2_regularization=0.1,
        random_state=42,
    )

    model_tmp.fit(X_train_tmp_processed, y_train_tmp)

    pred_log = model_tmp.predict(X_valid_tmp_processed)
    pred = np.clip(np.expm1(pred_log), 0, None)

    print(start_date)
    print("RMSLE:", rmsle(y_valid, pred))
    print("pred / actual total:", pred.sum() / y_valid.sum())
    print()
```

```python
all_data["is_promo"] = (all_data["onpromotion"] > 0).astype(int)
all_data["log_onpromotion"] = np.log1p(all_data["onpromotion"])

promo_stats = (
    train[train["date"] < valid_start]
    .groupby(["store_nbr", "family"])["onpromotion"]
    .agg(
        store_family_promo_mean="mean",
        store_family_promo_max="max",
    )
    .reset_index()
)

all_data = all_data.merge(
    promo_stats,
    on=["store_nbr", "family"],
    how="left",
)

all_data["promo_vs_mean"] = all_data["onpromotion"] / (
    all_data["store_family_promo_mean"] + 1e-6
)

all_data["promo_vs_max"] = all_data["onpromotion"] / (
    all_data["store_family_promo_max"] + 1e-6
)

all_data["is_payday"] = (
    (all_data["day"] == 15) |
    (all_data["is_month_end"] == 1)
).astype(int)

all_data["is_payday_window"] = (
    all_data["day"].isin([14, 15, 16]) |
    all_data["date"].dt.is_month_end |
    (all_data["date"] + pd.Timedelta(days=1)).dt.is_month_end |
    (all_data["date"] - pd.Timedelta(days=1)).dt.is_month_end
).astype(int)


group_cols = ["store_nbr", "family"]

for lag in [1, 7, 14, 28]:
    all_data[f"lag_{lag}"] = (
        all_data
        .groupby(group_cols)["sales"]
        .shift(lag)
    )

for window in [7, 14, 28, 56, 112]:
    all_data[f"rolling_mean_{window}"] = (
        all_data
        .groupby(group_cols)["sales"]
        .transform(lambda s: s.shift(1).rolling(window).mean())
    )

    all_data[f"rolling_std_{window}"] = (
        all_data
        .groupby(group_cols)["sales"]
        .transform(lambda s: s.shift(1).rolling(window).std())
    )

eps = 1e-6

all_data["trend_7_28"] = all_data["rolling_mean_7"] / (all_data["rolling_mean_28"] + eps)
all_data["trend_14_28"] = all_data["rolling_mean_14"] / (all_data["rolling_mean_28"] + eps)
all_data["trend_7_56"] = all_data["rolling_mean_7"] / (all_data["rolling_mean_56"] + eps)
all_data["trend_28_112"] = all_data["rolling_mean_28"] / (all_data["rolling_mean_112"] + eps)

model_train_start = pd.Timestamp("2016-08-01")

train_model_recent = train_model[
    train_model["date"] >= model_train_start
].copy()

X_train = train_model_recent[feature_cols]
y_train = np.log1p(train_model_recent["sales"])

X_valid = valid_model[feature_cols]
y_valid = valid_model["sales"]

print(train_model_recent.shape)
print(train_model_recent["date"].min(), train_model_recent["date"].max())

drop_cols = ["id", "date", "sales", "is_train"]

feature_cols = [col for col in train_model.columns if col not in drop_cols]

cat_cols = train_model[feature_cols].select_dtypes(
    include=["object", "string"]
).columns.tolist()

num_cols = [col for col in feature_cols if col not in cat_cols]

print("categorical:", cat_cols)
print("numeric:", len(num_cols))
```

```python
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        (
            "encoder",
            OrdinalEncoder(
                handle_unknown="use_encoded_value",
                unknown_value=-1,
            ),
        ),
    ]
)

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ]
)

X_train_processed = preprocess.fit_transform(X_train)
X_valid_processed = preprocess.transform(X_valid)

print("processed train:", X_train_processed.shape)
print("processed valid:", X_valid_processed.shape)

model = HistGradientBoostingRegressor(
    max_iter=1,
    learning_rate=0.06,
    max_leaf_nodes=31,
    l2_regularization=0.1,
    random_state=42,
    warm_start=True,
)

total_iter = 300
step = 10

for n_iter in tqdm(range(step, total_iter + step, step)):
    model.set_params(max_iter=n_iter)
    model.fit(X_train_processed, y_train)

valid_pred_log = model.predict(X_valid_processed)
valid_pred = np.expm1(valid_pred_log)
valid_pred = np.clip(valid_pred, 0, None)

score = rmsle(y_valid, valid_pred)
mae = mean_absolute_error(y_valid, valid_pred)

print("Model RMSLE:", score)
print("Model MAE:", mae)
```

```python
leaky_cols = [
    "lag_1",
    "lag_7",
    "lag_14",
    "rolling_mean_7",
    "rolling_std_7",
    "rolling_mean_14",
    "rolling_std_14",
    "trend_7_28",
    "trend_14_28",
    "trend_7_56",
]

safe_feature_cols = [
    col for col in feature_cols
    if col not in leaky_cols
]

safe_cat_cols = train_model[safe_feature_cols].select_dtypes(
    include=["object", "string"]
).columns.tolist()

safe_num_cols = [col for col in safe_feature_cols if col not in safe_cat_cols]

print("removed:", leaky_cols)
print("safe features:", len(safe_feature_cols))
print("safe categorical:", safe_cat_cols)
print("safe numeric:", len(safe_num_cols))
```
